**Código para importar arquivo PDF para o Google Drive**

```python
//Importa o Ficheiro
import requests

url = "https://d1.amobbs.com/bbs_upload782111/files_31/ourdev_569653.pdf"
r = requests.get(url)

with open("/content/arquivo.pdf", "wb") as f:
    f.write(r.content)
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
import shutil

shutil.copy("/content/arquivo.pdf", "/content/drive/MyDrive/arquivo.pdf")
```
